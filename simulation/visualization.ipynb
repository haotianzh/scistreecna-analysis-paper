```python
import pandas as pd 
import seaborn as sns 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# simulated from cellcoal
df = pd.read_csv('./cellcoal_simulation_results_all.csv')
color_palette = ['#1B9E77', '#D95F02', '#7570B3', '#66A61E', '#E7298A']
df1 = df[(df['metric'] == 'tree_acc')]
df2 = df[(df['metric'] == 'geno_acc')]
fig, axes = plt.subplots(figsize=(12, 4), nrows=1, ncols=2)
sns.boxplot(df1, ax=axes[0], x='dataset', y='value', hue='method', showmeans=True, showfliers=False, meanprops={"marker":"^","markerfacecolor":"white"}, palette=color_palette)
sns.boxplot(df2, ax=axes[1], x='dataset', y='value', hue='method', showmeans=True, showfliers=False, meanprops={"marker":"^","markerfacecolor":"white"}, palette=color_palette[-3:], width=0.5)
axes[0].set_xticklabels(['50', '100', '150','200'])
axes[1].set_xticklabels(['50', '100', '150','200'])
axes[0].set_xlabel('#Cell (#Site=#Cell)', fontsize=12)
axes[1].set_xlabel('#Cell (#Site=#Cell)', fontsize=12)
axes[0].set_ylabel('Tree Accuracy', fontsize=12)
axes[1].set_ylabel('Genotype Accuracy', fontsize=12)
axes[0].legend_.remove()
axes[1].legend_.remove()
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(loc='lower center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, shadow=True, ncol=5, handles=handles, labels=['Neighbor Joining (CN)',  'DICE-star', 'ScisTree2', 'CellPhy', 'ScisTreeCNA'])
fig.suptitle('Comparison of methods on CellCoal-simulated data', fontsize=18)
for i in range(0, 5):
    if i % 2 == 1:
        axes[0].axvspan(i-.5, i+.5, facecolor='black', alpha=0.1)
        axes[1].axvspan(i-.5, i+.5, facecolor='black', alpha=0.1)
plt.tight_layout()
plt.savefig('figures/sim_cellcoal.png', dpi=500, bbox_inches = 'tight')
```

```python
# === Heatmap version — compact for double-column paper ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

def reformat(df):
    df = df.copy()
    df['index'] = df['dataset'].str.split('_').str[-1]
    df['dataset'] = '500_500'
    df.loc[df['method'] == 'scistreecna', 'method'] = 'scistreec'
    return df

# Load data
df_low = pd.read_csv('./scsim_simulation_results_low_coverage_high_cn_noise.csv')
df_low_200 = pd.read_csv('./scsim_simulation_results_low_coverage_high_cn_noise_n200.csv')
df_low_500 = pd.read_csv('/data/haotian/snvcnv/other_phylogeny/simulation_scsim_clt_500_low/results.csv')
df_high = pd.read_csv('./scsim_simulation_results.csv')
df_high_500 = pd.read_csv('./other_phylogeny/simulation_scsim_clt_500/results.csv')

df_low_all = pd.concat([df_low, df_low_200, reformat(df_low_500)])
df_low_all['coverage'] = 'Low (1x)'
df_high_all = pd.concat([df_high, reformat(df_high_500)])
df_high_all['coverage'] = 'High (10x)'
df_all = pd.concat([df_low_all, df_high_all], ignore_index=True)

dataset_map = {'50_50': '50', '100_100': '100', '150_150': '150', '200_200': '200', '500_500': '500'}
df_all['n_cells'] = df_all['dataset'].map(dataset_map)

method_names = {'copy_nj': 'NJ (CN)', 'dice': 'DICE-star', 'scistree2': 'ScisTree2',
                'cellphy': 'CellPhy', 'scistreec': 'ScisTreeCNA'}
df_all['method_label'] = df_all['method'].map(method_names)

method_order = ['NJ (CN)', 'DICE-star', 'ScisTree2', 'CellPhy', 'ScisTreeCNA']
cell_order = ['50', '100', '150', '200', '500']
cov_order = ['Low (1x)', 'High (10x)']

# Compute mean values
means = df_all.groupby(['metric', 'coverage', 'method_label', 'n_cells'])['value'].mean().reset_index()

# Build pivot tables: rows = (coverage, method), cols = n_cells
fig = plt.figure(figsize=(3.5, 3.8))
gs = gridspec.GridSpec(2, 2, width_ratios=[1, 0.04], hspace=0.35, wspace=0.05)

for row, (metric, title) in enumerate([('tree_acc', 'Tree Accuracy'), ('geno_acc', 'Genotype Accuracy')]):
    sub = means[means['metric'] == metric]
    methods = method_order if metric == 'tree_acc' else ['ScisTree2', 'CellPhy', 'ScisTreeCNA']

    # Build matrix: rows grouped by coverage, cols by cell count
    row_labels = []
    matrix = []
    for cov in cov_order:
        for m in methods:
            row_labels.append(f'{m}')
            vals = []
            for c in cell_order:
                v = sub[(sub['coverage'] == cov) & (sub['method_label'] == m) & (sub['n_cells'] == c)]['value']
                vals.append(v.values[0] if len(v) > 0 else np.nan)
            matrix.append(vals)

    matrix = np.array(matrix)
    ax = fig.add_subplot(gs[row, 0])
    cax = fig.add_subplot(gs[row, 1])

    im = ax.imshow(matrix, aspect='auto', cmap='YlOrRd', vmin=np.nanmin(matrix), vmax=np.nanmax(matrix))

    # Annotate cells with values
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            v = matrix[i, j]
            if not np.isnan(v):
                color = 'white' if v > (np.nanmax(matrix) + np.nanmin(matrix)) / 2 else 'black'
                ax.text(j, i, f'{v:.2f}', ha='center', va='center', fontsize=5, color=color)

    # Draw separator between Low and High coverage
    sep = len(methods) - 0.5
    ax.axhline(sep, color='white', linewidth=2)

    ax.set_xticks(range(len(cell_order)))
    ax.set_xticklabels(cell_order if row == 1 else [''] * len(cell_order), fontsize=7)
    ax.set_yticks(range(len(row_labels)))
    ax.set_yticklabels(row_labels, fontsize=6)

    # Coverage group labels on right side
    n = len(methods)
    ax.text(len(cell_order) + 0.8, n / 2 - 0.5, 'Low\n(1x)', ha='center', va='center',
            fontsize=5.5, fontstyle='italic', transform=ax.transData)
    ax.text(len(cell_order) + 0.8, n + n / 2 - 0.5, 'High\n(10x)', ha='center', va='center',
            fontsize=5.5, fontstyle='italic', transform=ax.transData)

    ax.set_title(title, fontsize=8, pad=4)
    plt.colorbar(im, cax=cax)
    cax.tick_params(labelsize=5.5)

axes = fig.get_axes()
fig.text(0.45, 0.02, '#Cells (#Sites=#Cells)', ha='center', fontsize=8)

# plt.savefig('figures/compact_comparison_heatmap.pdf', bbox_inches='tight', dpi=300)
plt.show()
```
